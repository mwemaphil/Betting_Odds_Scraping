```python
import requests
import pandas as pd
import numpy as np
import json
```

```python
#Scraping Odds Data from Sportpesa

url = 'https://www.ke.sportpesa.com/api/upcoming/games'
params = {
    'type': 'prematch',
    'sportId': 1,
    'section': 'upcoming',
    'markets_layout': 'multiple',
    'o': 'leagues',
    'pag_count': 150,
    'pag_min': 16 
}
headers = {
    'Accept': 'application/json, text/plain, /',
    'Accept-Language': 'en-GB,en-US;q=0.9,en;q=0.8',
    'Connection': 'keep-alive',
    'Referer': 'https://www.ke.sportpesa.com/en/sports-betting/football-1/upcoming-games/?filterDay=-1&paginationOffset=45',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-origin',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36',
    'X-App-Timezone': 'Africa/Nairobi',
    'X-Requested-With': 'XMLHttpRequest',
    'sec-ch-ua': '"Google Chrome";v="111", "Not(A:Brand";v="8", "Chromium";v="111"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"'
}

res = requests.get(url, params=params, headers=headers)


dataframe = res.json()
print(dataframe)



#header = res.headers['Content-type'] 
#print(header)
#if 'application/json' in header:
#    event = res.json()[0]
#    print(event['date'])
#    print(len(res.json()))
#else:
#    print('Not a valid JSON')

```
